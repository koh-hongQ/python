```python
import pandas as pd

import seaborn as sns

targetUrl_iris="https://raw.githubusercontent.com/koh-hongQ/python/main/caffeine.csv"

irisData = pd.read_csv(targetUrl_iris)

print(irisData.shape)

print(irisData.head())

irisData.tail()
```

```python
sns.set_theme()
sns.set(rc={"figure.dpi":300}) #이미지 품질 설정
sns.set(rc={"figure.figsize":(6,3)}) ## 크기설정


## Scatter plot 산점도

sns.scatterplot(x="Volume (ml)", y ="Calories", data=irisData, hue ='type')
```

```python
## 히스토그램 histogram

sns.histplot( y= "Volume (ml)", data = irisData)
```

```python
sns.histplot( x= "Volume (ml)", data = irisData)
```

```python
sns.histplot( x= "Volume (ml)", data = irisData, kde = "True")
```

```python
sns.histplot( x= "Volume (ml)", data = irisData, hue = "type")
```

```python
## Barplot : 숫자 변수의 경향 추청치를 보여준다.

sns.barplot(x="type", y="Volume (ml)", data =irisData)
```

```python
## Box plot

sns.boxplot(x="type", y="Volume (ml)", data=irisData)
```
